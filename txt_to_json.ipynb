```python
import json
import os
```

```python
working_folder = r'C:\Users\leohernandez\OneDrive - Deloitte (O365D)\Documents\Proyectos\DttLab\llama-stack-apps\llm-rag-files\all_accounts'
all_accounts_file = f'{working_folder}\\accounts.txt'
output_file = f'{working_folder}\\output.json'
```

```python
def find_depth(line: str) -> int:
    depth = 1
    symbol = '\t'
    for char in line:
        if char.isalpha():
            break
        if char == symbol:
            depth += 1
    return depth
```

```python
file = open(file=all_accounts_file, mode="r", encoding='UTF-8')
def create_nodes(text):
    lines = text.splitlines()
    stack = []
    result = []
    id_counter = 1

    for line in lines:
        if not line.strip():
            continue

        nivel = find_depth(line)
        nombre = line.strip()

        node = {"id": id_counter,
                "nombre": nombre,
                "nivel": nivel,
                "hijos": []}
        
        id_counter += 1

        # Determinar dónde insertar el nodo en la jerarquía
        if nivel == 1:
            path = f'{nombre}'
            node['path'] = path
            result.append(node)
            stack.clear()
            stack.append((nivel, node, path))
        else:
            # Limpiar el stack hasta encontrar el nivel correcto
            while stack and stack[-1][0] >= nivel:
                stack.pop()

            path = f'{stack[-1][2]} > {nombre}'
            if stack:
                # Insertar el nodo como hijo del último nodo válido en el stack
                node['path'] = path
                stack[-1][1]["hijos"].append(node)

            stack.append((nivel, node, path))

    return result

# Texto de entrada con jerarquía de tabulaciones
text = file.read()

# Generar el JSON jerárquico
account_hierarchy = create_nodes(text)

with open(output_file, 'w', encoding='UTF-8') as fp:
    json.dump(account_hierarchy, fp, ensure_ascii=False)
```
